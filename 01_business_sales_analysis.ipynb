# SME Sales Analytics & Revenue Forecasting

This notebook performs a business analysis of SME retail sales data.

## 1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

DATA_PATH = Path('../data/sme_sales_data.csv')
```

## 2. Load Data

```python
df = pd.read_csv(DATA_PATH)
df['order_date'] = pd.to_datetime(df['order_date'])
df.head()
```

## 3. Business KPIs

```python
total_revenue = df['revenue'].sum()
total_profit = df['profit'].sum()
orders = df['invoice_id'].nunique()
customers = df['customer_id'].nunique()
aov = total_revenue / orders
profit_margin = total_profit / total_revenue

pd.DataFrame({
    'Metric': ['Revenue', 'Profit', 'Orders', 'Customers', 'Average Order Value', 'Profit Margin'],
    'Value': [total_revenue, total_profit, orders, customers, aov, profit_margin]
})
```

## 4. Monthly Revenue Trend

```python
monthly = df.groupby(pd.Grouper(key='order_date', freq='ME')).agg(
    revenue=('revenue', 'sum'),
    profit=('profit', 'sum'),
    orders=('invoice_id', 'nunique')
).reset_index()

plt.figure(figsize=(10, 5))
plt.plot(monthly['order_date'], monthly['revenue'], marker='o')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 5. Category Performance

```python
category_perf = df.groupby('category').agg(
    revenue=('revenue', 'sum'),
    profit=('profit', 'sum'),
    units_sold=('quantity', 'sum')
).sort_values('revenue', ascending=False)
category_perf
```

## 6. Business Recommendations

- Focus on high-profit categories, not only high-revenue categories.
- Use customer segmentation for targeted promotions.
- Monitor monthly revenue trends for inventory planning.
- Compare sales channels to decide where marketing budget should go.
- Track discount levels to avoid damaging profit margins.
